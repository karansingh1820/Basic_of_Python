```python
mylist1=['Delhi','Mumbai', 'Chennai', 'Kolkata']
mylist2=[40,10,42,30,40,32,100,40,90]
mylist3=['HYD','Banglore']

#printing mylist1
print ("The list1=",mylist1)
print("\n")

#printing mylist3
print ("The list3=",mylist3)
print("\n")

#Converting list into string-str()
print("Converted list into String=",str(mylist1))
strm=str(mylist1)
print("STring Index values:", strm[0])
print("STring Index values:", strm[1])
print("STring Index values:", strm[2])



#concatenation of two list
print("Merging or Concatenation-Mylist1+Mylist3",mylist1+mylist3)

print("\n")

#storing two list in another list after concantenation
mylist4=mylist1+mylist3
print("After Merging or Concatenation and storing in Different list:")
print(mylist4)

print("\n")

#finding total Elements in a list
print ('The total elelements in list1=',len(mylist1))

print("\n")

#find length of list1, list2, list3 and also find whole length(length of all list)
len_list1=len(mylist1)
len_list2=len(mylist2)
len_list3=len(mylist3)
print("The Length of List1=",len_list1)
print("The Length of List2=",len_list2)
print("The Length of List3=",len_list3)
len_total=len_list1+len_list2+len_list3
print("The Length of All list=",len_total)

print("\n")

#updating list- List works on index so we need to consider index of an element in a list for accessing that element.
#Here I need to change Mumbai with Chicago
mylist1[1]="Chicago"
print("After Update the list1=",mylist1)


print("\n")
#Finding Maximum element in a list2
print( "The Max Element in list2=",max(mylist2))

print("\n")
#Finding minimum element in a list2
print( "The Min Element in list2=",min(mylist2))

print("\n")
#Append-Inserting an element at end of a list is called append.
mylist1.append("Gurgaon")
print("After Append the List1=",mylist1)


print("\n")
#Count-It will count the occurance of an item in a list.
print("The Occurance of 40 in list2=",mylist2.count(40))


print("\n")
#Extend-It will concatenate a  list(mylist3) in a list(mylist1) at end.
mylist1.extend(mylist3) # it is equivalent to mylist1=mylist1+mylist3
print("The Exended list1=",mylist1)

print("\n")
#index-It is used for checking existence of an element in list, if item not found then it will generate an exception ValueError. Need to manage exception					
try:
    print("The Item(Chennai) found at the index=",mylist1.index("Chennai"))
except:
    print("Item not Found in list:ValueError")



print("\n")    
#Insert-It is used to insert an element at any position with exiting indices.
mylist1.insert(3, 'Chandigarh')

print("After insert at index 3 the list=",mylist1)


print("\n")
#Remove-It will delete an element from a list
mylist1.remove('Chandigarh')
print("After Remove of Chandigarh the list=", mylist1)


print("\n")
#Pop-It mean delete elements from last(-) and from begnning(+).  
#Example with -(Delete from end)
item=mylist1.pop(-2)       #It will delete element from right at index -2 and store back the element item
print("The removed item with pop=",item)
print("After pop the list =",mylist1)


print("\n")
#Example with +(Delete from begining)
item=mylist1.pop(2)              #It will delete element from left at index 2 and store back the element item
print("The removed item with pop=",item)
print("After pop the list =",mylist1)


print("\n")

#Reverse-It will reverse the element of list
mylist1.reverse()
print("The Reverse list =",mylist1)

print("\n")
#sort-It arrange the data in ascending or descending order
#Example for Ascending order
mylist2.sort()
print("The ascending order sorted list =",mylist2)

print("\n")
#Example for descending order
mylist2.sort(reverse=True)              #//Descending order sorting
print("The Descening order sorted list =",mylist2)

print("\n")
#deleting an element.
del mylist1[1]
print("After Deleting an Element the list1=",mylist1)

print("\n")
#deleting multiple elements
del mylist1[1:3]
print("After Deleting multiple Elements the list1=",mylist1)

print("\n")
#clear the list-means removing  all elements only but list remain empty in the memory
mylist1.clear()
print("After Clear the List1=",mylist1)

print("\n")
#Deleting entire list- if list doesnot exist, will gender exception if you access.
del mylist1
print(mylist1)
```

```python
# Slicing- a string with slice function-
String = 'ASTRING'

# Using slice constructor
s1 = slice(3)            # Creates a slice object that takes the first 3 elements
s2 = slice(1, 5, 2)     # Creates a slice object that takes elements from index 1 to 4, stepping by 2
s3 = slice(-1, -12, -2) # Creates a slice object that takes elements from index -1 to -12, stepping by -2

print("String slicing:--->")
print(String[s1])       # Output: 'AST'
print(String[s2])       # Output: 'SRI'
print(String[s3])       # Output: 'GNR'

```
