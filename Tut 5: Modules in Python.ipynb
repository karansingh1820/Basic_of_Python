```python
import math
result1=math.sqrt(120)
print("Squre root:",result1)

result2=math.pow(10,5)
print("Power value:",result2)

result3=math.floor(102.5453)                          #Decrease the value by nearest value, it will ignore point
print("Floor Value:",result3)  

result4=math.ceil(102.5453)                         # Increase by nearest value
print("Ceil Value:",result4)
#Output: 103


result5=round(102.545646464643455353535,2)        # it is not a math  function
print("Round value:",result5)                   #Output:102.55	


result6= math.log(10)
print("The Log value:", result6)
result6= math.log10(30)
print("The Log10 value:", result6)


num=-230.0
print("The Absolute value :",math.fabs(num))  #it is given math module
print("The Absolute value :",abs(num))  #it is given python main module
```

```python
# How to calcualte random %

import random
print(round(random.random()*100,2),"%")



# This code generates a random percentage with two decimal places:

# random.random() generates a random floating-point number between 0.0 and 1.0.
# Multiplying by 100 scales this value to a range between 0.0 and 100.0.
# round(..., 2) rounds the resulting number to two decimal places.
# The print function then displays this number followed by a percentage sign (%).
# For example, if random.random() returns 0.657, the code will output 65.70%.
```
